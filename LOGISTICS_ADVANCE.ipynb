```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("C:/Users/akash/Downloads/logistics_dataset.csv")
print(df)
```

```python
# let's check the null values 
# and data type and also check the all information about this data set 
# and then we will create a charts , bars  for beter understanding 
# easy to visualizes 
```

```python
df.info() # checking the  information about data and their type 
```

```python
df.describe() # describe the data 
```

```python
numeric_df = df.select_dtypes(include = 'number')


numeric_df.corr() # correlation btw weight and cost means cost per weight 
```

```python
# we clearly see our correlation is not positive it's mean cost per 
# weight is very low and we are in loss we have to increase cost per weight 
```

```python
# now we create a heatmap for better visualization and understanding 
correlation = numeric_df.corr()

sns.heatmap(correlation,annot=True,cmap="coolwarm" , linewidths = 0.5)
plt.title("CORRELATION BTW SHIPMENT WEIGHT AND DELIVERY COST")
plt.show()
```

```python
# now we create a boxplot for visualizing data 

plt.figure(figsize=(8,5))
sns.boxplot(x =df['delivery_cost'],color = 'skyblue')
plt.title('BOXPLOT OF DELIVERY COST',fontsize = 14)
plt.grid(True)
plt.show()
```

```python
# create scatterplot for visiable trend
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='shipment_weight',y='delivery_cost',color='purple',alpha=0.6)
plt.title('shipment weight vs delivery cost ',fontsize=14)
plt.xlabel('shipment weight (kg)')
plt.ylabel('delivery cost ($)')
plt.grid(True)
plt.show()
```

```python
# import histogram for delivery cost 

plt.figure(figsize=(8, 5))
sns.histplot(df['delivery_cost'], bins=30, kde=True, color='salmon')
plt.title("Distribution of Delivery Cost", fontsize=14)
plt.xlabel("Delivery Cost ($)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['shipment_weight'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Shipment Weight", fontsize=14)
plt.xlabel("Shipment Weight (kg)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
